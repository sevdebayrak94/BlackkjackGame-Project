```python
import random
cards = [11,2,3,4,5,6,7,8,9,10,10,10,10]
def random_card():
    """Returns a random card from the deck"""
    return random.choice(cards)

def calculate_score(list):
    if sum(list) == 21 and len(list) == 2:
        return 0 #this represent a Blackjack
    if 11 in list and sum(list)> 21:
        list.remove(11)
        list.append(1)
    return sum(list)

def compare(user_score, computer_score):
     if  user_score > 21 or computer_score > 21:
         return "You went over. You loose"
     elif user_score == computer_score :
         return "Draw"
     elif computer_score == 0:
         return "You lose"
     elif user_score == 0:
         return "You Win :))"
     elif user_score > 21 :
         return "You lost"
     elif computer_score > 21:
         return "Opponent went over. You win"
     elif user_score > computer_score :
         return "You win"
     else:
         return "You Lose"


def play_game():
    user_cards = []
    computers_cards = []
    computer_score = -1
    user_score = -1
    is_game_over = False
    #We want to chose 2 cards in the begining
    for i in range(2):
        user_cards.append(random_card())
        computers_cards.append(random_card())


    while not is_game_over:
        user_score = calculate_score(user_cards)
        computer_score = calculate_score(computers_cards)
        print(f"Your cards: {user_cards}, current_score: {user_score}")
        print(f"Computer's first card: {computers_cards[0]}")

        if user_score == 0 or computer_score == 0 or user_score > 21:
            is_game_over == True
        else:
            user_should_deal = input("Do you want to draw another card [y/n]: ")
            if user_should_deal == "y":
                user_cards.append(random_card())
                calculate_score(user_cards)
            else:
                is_game_over = True


    # Once the user is done, it is time to let the computer play. The computer should keep...
    # drawing cards as long as it has a score less than 17.
    while computer_score != 0 and computer_score < 17:
        computers_cards.append(random_card())
        computer_score = calculate_score(computers_cards)

    print(f"Your final card: {user_cards}' final score: {user_score}")
    print(f"Computer's cards: {computers_cards}' final score: {computer_score}")
    print(compare(user_score, computer_score))

while input("Do you want to play BlackJack [y/n]: ") == "y":
    play_game()
```
